```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow import keras
from tensorflow.keras import layers
```

### __1. Import and data pre-processing:__

```python
# import counts dataframe:

df_counts = pd.read_csv("../dataset/mdd_counts.csv") 
df_counts = df_counts.set_index("Unnamed: 0")

print(df_counts.shape)
df_counts.head()
```

```python
# import metadata:

df_meta = pd.read_csv("../dataset/mdd_meta.csv")
df_meta = df_meta.set_index("!Sample_title")
df_meta.head()
```

```python
# select (x, y) data:
# x -> counts table
# y -> (sample, phenotype, denger)

counts = df_counts.to_numpy()
print(counts.min(), counts.mean(), counts.max())

plt.figure(figsize=(15,3))
plt.imshow(counts)
plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# normalization:
x = np.log(counts+1)
x /= x.max()

print(x.min(), x.mean(), x.max())

plt.figure(figsize=(15,3))
plt.imshow(x)
plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
# meta -> (subject_id, sample, phenotype, gender)

meta = [x.split(".") for x in df_counts.index]

N = len(meta)

for i in range(N):
    n = meta[i][0]
    
    meta[i] += df_meta.loc[['phenotype', 'gender'], n].tolist()

meta = np.array(meta)
print(meta.shape)
meta
```

```python
# one-hot encoding:

def one_hot(meta):

    labels = np.array(list(set(meta)))
    hot = np.zeros((len(meta), len(labels)))

    for i, x in enumerate(meta):
        j = np.where(x == labels)[0]
        hot[i][j] = 1.0

    return hot

y1 = one_hot(meta.T[1])
y2 = one_hot(meta.T[2])
y3 = one_hot(meta.T[3])

print(f"y1:{y1.shape}, y2:{y2.shape}, y3:{y3.shape}")

y1[:2], y2[:2], y3[:2]
```

```python
# shuffling:
N_samples = x.shape[0]
i = np.random.permutation(N_samples)

x, y1, y2, y3 = x[i], y1[i], y2[i], y3[i]
meta = meta[i]

x.shape, y1.shape, y2.shape, y3.shape
```

```python
# splitting (x, y) into (train, val, test):
N_test = int(0.1*N_samples)
N_val = int(0.2*N_samples)
N_train = N_samples - (N_val + N_test)

x_train = x[:N_train]
x_val = x[N_train:N_train+N_val]
x_test = x[N_train+N_val:]

print(f"x-train:{x_train.shape}, x-val:{x_val.shape} x-test:{x_test.shape}")

y1_train = y1[:N_train]
y1_val = y1[N_train:N_train+N_val]
y1_test = y1[N_train+N_val:]

print(f"y1-train:{y1_train.shape}, y1-val:{y1_val.shape} y1-test:{y1_test.shape}")

y2_train = y2[:N_train]
y2_val = y2[N_train:N_train+N_val]
y2_test = y2[N_train+N_val:]

print(f"y2-train:{y2_train.shape}, y2-val:{y2_val.shape} y2-test:{y2_test.shape}")

y3_train = y3[:N_train]
y3_val = y3[N_train:N_train+N_val]
y3_test = y3[N_train+N_val:]

print(f"y3-train:{y3_train.shape}, y3-val:{y3_val.shape} y3-test:{y3_test.shape}")
```

```python
# deleting unnecessary arrays:
del df_counts, df_meta, counts
```

### __2. Neural network modeling:__

```python
In = keras.Input((x_train.shape[1], ))

x = layers.Dense(400, activation='relu')(In)

x = layers.Dense(200, activation='relu')(x)
x = layers.Dense(20, activation='relu')(x)
x = layers.Dense(200, activation='relu')(x)
x = layers.Dense(400, activation='relu')(x)

Out = layers.Dense(x_train.shape[1], activation='sigmoid')(x)

model = keras.Model(inputs=In, outputs=Out)
model.summary()
```

### __3. Model compilation:__

```python
model.compile(optimizer="Adam", loss="mse", metrics=['mae'])
```

### __4. Train and validation__   

### __5. Final training__    

```python
report = model.fit(x=np.concatenate([x_train, x_val]), 
                   y=np.concatenate([x_train, x_val]),
                   epochs=20, batch_size=1)
```

```python
loss = report.history['loss']
mae = report.history['mae']
epochs = np.linspace(1, len(loss), len(loss))

plt.figure(figsize=(7, 4))

plt.plot(epochs, loss, label="loss")
plt.plot(epochs, mae, label="mae")
plt.xlabel("epochs")
plt.legend()
plt.grid()

plt.show()
```

### __6. Test evaluation__:

```python
model.evaluate(x=x_test, y=x_test, batch_size=1)
```

```python
x = np.concatenate([x_train, x_val, x_test])

x_pred = model.predict(x)
x_pred.shape
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 6))

ax[0].imshow(x)
ax[1].imshow(x_pred)

for i in range(2):        
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    
plt.show()
```

#### __6.1 Anomaly detection:__

```python
mse = np.mean(pow(x-x_pred, 2), axis=1)
mse.shape
```

```python
threshold = 0.0010

plt.hist(mse, bins=50, histtype="step", lw=2)
plt.axvline(x=threshold, color='pink')
plt.xlabel("MSE")
plt.ylabel("Number of samples")
plt.show()
```

```python
j = np.where(mse >= threshold)[0]

meta[j]
```
